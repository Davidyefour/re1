# 灰度图，颜色相反

```python
# 灰度图计算方式：Grey = 0.299*R + 0.587*G + 0.114*B
# 优化后 Grey = （306*R+601*G + 116*B） >> 10

# 颜色反转： [255, 255, 255] - im (im = Image.open("dir")
```

```python
from PIL import Image
img = Image.open("./Tower.jpg")
```

```python
img
```

```python
import numpy as np
img_arr = np.array(img)
```

```python
img_arr.size
```

```python
b = [255, 255, 255]-img_arr
```

```python
b = b.astype("uint8")
```

```python
im = Image.fromarray(b)
```

```python
im
```

```python
im.save("canton2.jpg")
```
